## copy only the files containing "१.१ विषय परिचय" or "१.१ विषयपरिचय" into a new folder named manually

```python
import glob
import os
import shutil

def search_phrases():
    file_paths = glob.glob('*.txt')
    found_files = []

    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

            if "१.१ विषय परिचय" in text or "१.१ विषयपरिचय" in text:
                found_files.append(file_path)

    return found_files

def copy_files(found_files, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for file_path in found_files:
        file_name = os.path.basename(file_path)
        destination_path = os.path.join(destination_folder, file_name)
        shutil.copy(file_path, destination_path)
        print(f"Copied {file_name} to {destination_folder}.")

found_files = search_phrases()
count = len(found_files)

if found_files:
    destination_folder = "manually/"  # Specify the path to the new folder
    copy_files(found_files, destination_folder)
    print(f"Successfully copied {count} file(s) to {destination_folder}.")
else:
    print("No files contain the phrases.")
```

# print out the text in between "१.१ विषय परिचय" and "१.२"

```python
import glob
import re

def extract_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

        # Define the regular expression pattern to match the desired texts
        pattern = r'१\.१ (विषय परिचय|विषयपरिचय)(.*?)१\.२'

        # Find all matches of the pattern in the text
        matches = re.findall(pattern, text, re.DOTALL)

        return matches

def process_files(folder_path):
    file_paths = glob.glob(folder_path + '/*.txt')

    extracted_texts = []
    for file_path in file_paths:
        matches = extract_text(file_path)
        extracted_texts.extend(matches)

    return extracted_texts

folder_path = "manually/"  # Specify the path to the folder containing the text files
extracted_texts = process_files(folder_path)

if extracted_texts:
    print("Extracted texts:")
    for match in extracted_texts:
        print("Text 1:", match[0])
        print("Text 2:", match[1])
        print()
else:
    print("No texts were extracted.")
```

# print out the text in between "१.१ विषय परिचय" and "१.२" where the no of words in between is more than 5
It is done to avoid displaying from table of content

```python
import glob
import re

def extract_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

        # Define the regular expression pattern to match the desired texts
        pattern = r'१\.१ (विषय परिचय|विषयपरिचय)(.*?)१\.२'

        # Find all matches of the pattern in the text
        matches = re.findall(pattern, text, re.DOTALL)

        extracted_texts = []
        for match in matches:
            text_between = match[1].strip()
            words_count = len(re.findall(r'\b\w+\b', text_between))

            if words_count > 5:
                extracted_texts.append(match)

        return extracted_texts

def process_files(folder_path):
    file_paths = glob.glob(folder_path + '/*.txt')

    extracted_texts = []
    for file_path in file_paths:
        matches = extract_text(file_path)
        extracted_texts.extend(matches)

    return extracted_texts

folder_path = "manually/"  # Specify the path to the folder containing the text files
extracted_texts = process_files(folder_path)

if extracted_texts:
    print("Extracted texts:")
    for match in extracted_texts:
        print("Text 1:", match[0])
        print("Text 2:", match[1])
        print()
else:
    print("No texts were extracted.")
```

# to store the file name with extracted text in csv file or something like that

```python
import csv
import glob
import re

def extract_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()

        # Define the regular expression pattern to match the desired texts
        pattern = r'१\.१ (विषय परिचय|विषयपरिचय)(.*?)१\.२'

        # Find all matches of the pattern in the text
        matches = re.findall(pattern, text, re.DOTALL)

        extracted_texts = []
        for match in matches:
            text_between = match[1].strip()
            words_count = len(re.findall(r'\b\w+\b', text_between))

            if words_count > 5:
                extracted_texts.append(match[1])

        return extracted_texts

def process_files(folder_path):
    file_paths = glob.glob(folder_path + '/*.txt')

    extracted_data = []
    for file_path in file_paths:
        file_name = file_path.split('/')[-1]
        extracted_texts = extract_text(file_path)
        extracted_data.append([file_name] + extracted_texts)

    return extracted_data

folder_path = "manually/"  # Specify the path to the folder containing the text files
extracted_data = process_files(folder_path)

if extracted_data:
    csv_file = "extracted_data.csv"  # Specify the name of the CSV file
    with open(csv_file, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['File Name', 'Extracted Text'])
        writer.writerows(extracted_data)

    print(f"CSV file '{csv_file}' created successfully.")
else:
    print("No texts were extracted.")
```
